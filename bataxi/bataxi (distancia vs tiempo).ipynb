```python
url = 'https://raw.githubusercontent.com/aaizemberg/vis/gh-pages/bataxi/dd.tsv'
```

```python
import pandas as pd
df = pd.read_csv(url, sep='\t')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
x = df.distancia / 1000
y = df.duracion / 60
plt.plot(x, y, '.', alpha=0.1)
plt.show()
```

```python
import numpy as np
m, b = np.polyfit(x, y, 1)
plt.plot(x, y, '.', alpha=0.1)
plt.plot(x, m*x + b, '-')
plt.show()
```

```python
#
# taxistas con viajes terminados en 250 mts, en 3 meses
#
df1 = df[['id_taxista','distancia']]
df1[df1.distancia < 300].groupby(['id_taxista']).count().sort_values(by='distancia', ascending=False).head(10)
```
