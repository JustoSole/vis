```
import pandas as pd

df = pd.read_json('https://interactive.twitter.com/brazil2014/data/json/area.json')
df.head()
```

```
import numpy as np
import matplotlib.pyplot as plt

x  = df.date
y1 = df.AecioNeves
y2 = df.dilmabr
y3 = df.eduardocampos40
y4 = df.silva_marina

plt.stackplot(x,y1, y2, y3, y4, labels=['Aecio Neves','Dilma','Eduardo Campos','Silva Marina'])
plt.legend(loc='upper left')
plt.show()
```

```
url = 'https://raw.githubusercontent.com/aaizemberg/vis/gh-pages/brasil/seguidores.tsv'

df2 = pd.read_csv( url, sep = '\t')
df2.head()

x  = df2.date
y1 = df2.p_AecioNeves
y2 = df2.p_dilmabr
y3 = df2.p_eduardocampos40
y4 = df2.p_silva_marina

plt.stackplot(x,y1, y2, y3, y4, labels=['Aecio Neves','Dilma','Eduardo Campos','Silva Marina'])
# plt.legend(loc='upper right')
plt.show()
```
