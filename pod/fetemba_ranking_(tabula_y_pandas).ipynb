```
# url = "https://raw.githubusercontent.com/aaizemberg/vis/gh-pages/pod/fetemba-julio16-2018.tsv"
# url = "https://raw.githubusercontent.com/aaizemberg/vis/gh-pages/pod/fetemba-2018-10-09.tsv"
url = "https://raw.githubusercontent.com/aaizemberg/vis/gh-pages/pod/fetemba-2019-03-21.tsv"

import pandas as pd
df = pd.read_csv(url, sep='\t')
df.head(10)
```

# Los mejores 10 clubes, segun el desempeño de sus jugadores

```
df[['club','variacion']].groupby('club').sum().sort_values(by='variacion', ascending=False).reset_index().head(10)
```

# Los peores 10 clubes, segun el desempeño de sus jugadores

```
df[['club','variacion']].groupby('club').sum().sort_values(by='variacion', ascending=True).reset_index().head(10)
```

# Los 10 jugadores que sumaron mas puntos

```
df[['nombre','categoria','club','variacion']].sort_values(by='variacion', ascending=False).head(10)
```

```
import altair as alt

# clubes que quiero comparar
#
comparar = ['Fenix', 'Alvear Club', 'Top Minc', 'Ferrocarril Oeste', 'Atlanta', 'La Reja', 'Inst. Compañía María', 'River Plate'] 
#            'Ntra. Señora Lourdes', 'La Legión TDM', 
#            'UGAB', 'Atlanta', 'Gimnasty', 'CTDM Pato Nizetich', 
#            'Soc. Fomento CAVA', 'La Reja', 'TMR', 'Macabi', 'San Lorenzo', 
#            'Peretz','Nikkei TDM']

df1 = df[df.club.isin(comparar)]

alt.Chart(df1).mark_circle(size=100, opacity=0.5).properties(
    title='Fetemba 2019/03/21').interactive().encode(
    x = 'variacion:Q',
    y = 'club:N',
    color = 'categoria:N', 
    tooltip = [ 'pos', 'nombre', 'categoria', 'club', 'puntos', 'variacion']
)
```

```
base = alt.Chart(df)

lower_box = 'q1(variacion):Q'
lower_whisker = 'min(variacion):Q'
upper_box = 'q3(variacion):Q'
upper_whisker = 'max(variacion):Q'

lower_plot = base.mark_rule().encode(
    x=alt.X(lower_whisker, title="variacion"),
    x2=lower_box,
    y='club:N'
)

middle_plot = base.mark_bar(size=15.0).encode(
    x=lower_box,
    x2=upper_box,
    y='club:N'
)

upper_plot = base.mark_rule().encode(
    x=upper_whisker,
    x2=upper_box,
    y='club:N'
)

middle_tick = base.mark_tick(
    color='white',
    size=14.0
).encode(
    x='median(variacion):Q',
    y='club:N'
)

lower_plot + middle_plot + upper_plot + middle_tick
```
